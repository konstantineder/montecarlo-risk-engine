## Basket CDS valuation

### 1 Theory

\begin{equation}
c_{CDS}(T)=\frac{(1-R)\int_{0}^TP(0,t)\tilde{\lambda}(t)S(t)\,\mathrm{d}t}{\sum_{i=1}^n P(0,t_i)S(t_i)}
\end{equation}

Assuming piecewise constant $\lambda$ we have $\tilde{\lambda}=\lambda$
\begin{equation}
c_{CDS}(T)=\frac{(1-R)\int_{0}^TP(0,t)\tilde{\lambda}(t)S(t)\,\mathrm{d}t}{\sum_{i=1}^n P(0,t_i)S(t_i)}
\end{equation}

\begin{equation}
S(t):=\mathbb{E}\left(\exp\left(-\int_0^t \lambda(s)\,\mathrm{d}s\right)\right)=:\exp\left(-\int_0^t \tilde{\lambda}(s)\,\mathrm{d}s\right)
\end{equation}

### 2 Bootstrapping hazard rates

$c_i:=c_{CDS}(T_i)$, $p_i:=S(t_i)$
\begin{equation}
\frac{c_i}{1-R}\sum_{k=1}^n P(0,t_k)p_k=\sum_{l=1}^i P(0,T_l)p_{l-1}-\sum_{l=1}^i P(0,T_l)p_l
\end{equation}

\begin{equation}
p_i = \frac{\sum_{l=1}^i P(0,T_l)p_{l-1}-\sum_{l=1}^{i-1} P(0,T_l)p_l-\frac{c_i}{1-R}\sum_{k=1}^{n-1} P(0,t_k)p_k}{P(0,T_i)\left(1+\frac{c_i}{1-R}\right)}
\end{equation}

```python
import numpy as np
from typing import List, Tuple

def bisection_search(func, low: float = 1e-10, high: float = 5.0,
                     tolerance: float = 1e-12, iters: int = 100) -> float:
    value_low, value_high = func(low), func(high)
    cnt = 0
    while value_low * value_high > 0.0 and cnt < 20:
        high *= 2.0
        value_high = func(high)
        cnt += 1
    if value_low * value_high > 0.0:
        raise RuntimeError("Could not bracket hazard in bucket")
    for _ in range(iters):
        mid = 0.5 * (low + high)
        value_mid = func(mid)
        if abs(value_mid) < tolerance or (high - low) < 1e-12:
            return mid
        if value_low * value_mid <= 0.0:
            high, value_high = mid, value_mid
        else:
            low, value_low = mid, value_mid
    return 0.5 * (low + high)


def compute_cds_legs(
    maturities: List[float],                     # T_i (years), ascending
    payment_days: np.ndarray,                   # coupon dates t_k (years), ascending
    discount_factors_payment_days: np.ndarray,  
    recovery_rate: float,
    hazard_rates: List[float]
) -> Tuple[float, float]:
    indices = np.searchsorted(payment_days, maturities)
    time_to_index = dict(zip(maturities, indices))

    deltas = [payment_days[0]] + [payment_days[k]-payment_days[k-1]
                                  for k in range(1, len(payment_days))]
    premium_leg = 0.0
    protection_leg  = 0.0
    survival_prob_prev = 1.0
    prev_time_idx = 0
    for idx, maturity in enumerate(maturities):
        prev_maturity = maturities[idx - 1] if idx > 0 else 0.0
        time_idx = time_to_index[maturity]
        hazard_rate = hazard_rates[idx]
        for k in range(prev_time_idx, time_idx + 1):  # include coupon at t_k = T_i
            payment_date = payment_days[k]
            discount = discount_factors_payment_days[k]
            delta = deltas[k]
            survival_prob = survival_prob_prev * np.exp(-hazard_rate * (payment_date - prev_maturity))
            # premium leg: coupons + AoD (trapezoid)
            accrual = 0.5 * delta * discount * (survival_prob_prev - survival_prob)
            premium_leg += delta * discount * survival_prob + accrual
            # protection leg
            protection_leg += (1.0 - recovery_rate) * discount * (survival_prob_prev - survival_prob)
        prev_time_idx = time_idx
        survival_prob_prev = survival_prob
    return premium_leg, protection_leg


def bootstrap_hazards(
    credit_spreads: List[float],                 
    maturities: np.ndarray,                     
    payment_days: np.ndarray,                   
    discount_factors_payment_days: np.ndarray,  
    recovery_rate: float                         
) -> List[float]:
    """
    Generalized CDS bootstrap (coupon grid + AoD trapezoid) with piecewise-constant hazards.
    Calls `compute_legs` during root-finding.
    Assumes each maturity T_i is present in `payment_days`.
    """
    assert len(payment_days) == len(discount_factors_payment_days)
    hazard_rates: List[float] = []

    for i, spread in enumerate(credit_spreads):
        # Build F_i(lambda_i) = spread * RPV01(T_i) - Prot(T_i),
        # where RPV01/Prot are computed by compute_legs on data up to T_i
        def F(lam_i: float) -> float:
            tmp_haz = hazard_rates + [lam_i]          
            mats_up_to_i = maturities[: i + 1]         
            prem, prot = compute_cds_legs(
                mats_up_to_i,
                payment_days,
                discount_factors_payment_days,
                recovery_rate,
                tmp_haz
            )
            return spread * prem - prot

        lam_i = bisection_search(F) 
        hazard_rates.append(lam_i)

    return hazard_rates

        
```

### 3 The Copula

```python
from scipy.stats import norm
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
FLOAT = torch.float32 if device.type == "cuda" else torch.float64

def gaussian_copula_unif(
    dim: int,
    num_samples: int,
    cov_matrix: torch.Tensor,
    seed: int | None = None,
) -> torch.Tensor:
    """
    Gaussian copula with antithetic variates.

    Returns:
        U: (num_samples, dim) uniforms in (0,1)
    """

    if num_samples % 2 != 0:
        raise ValueError("num_samples must be even when using antithetic variates")

    # RNG
    g = torch.Generator(device=device)
    if seed is not None:
        g.manual_seed(seed)

    # Cholesky of covariance matrix
    chol = torch.linalg.cholesky(cov_matrix)

    # iid standard normals (half)
    z = torch.randn(
        num_samples // 2,
        dim,
        generator=g,
        device=device,
        dtype=FLOAT,
    )

    # antithetic normals
    z_a = torch.cat([z, -z], dim=0)      

    # correlate
    corr_normals = z_a @ chol.T           

    # standard normal CDF
    U = 0.5 * (1.0 + torch.erf(corr_normals / np.sqrt(2.0)))

    # numerical safety: keep strictly inside (0,1)
    eps = torch.finfo(FLOAT).eps
    U = U.clamp(min=eps, max=1.0 - eps)

    return U


def uniforms_to_default_times(
    uniforms: torch.Tensor,
    hazards: torch.Tensor,
    maturities: torch.Tensor
) -> torch.Tensor:
    """
    Compute correlated default times based on correlated uniforms.

    Args:
        uniforms:   (num_samples, num_names) ~ U(0,1)
        hazards:    (num_names, num_buckets) piecewise-constant lambdas
        maturities: (num_buckets,) ascending bucket endpoints

    Returns:
        tau: (num_samples, num_names) default times
             If no default by last bucket, sets tau = maturities[-1].
    """
    num_samples, num_names = uniforms.shape
    num_names_h, num_buckets = hazards.shape
    if num_names != num_names_h:
        raise ValueError("uniforms.shape[1] must match hazards.shape[0]")
    if maturities.numel() != num_buckets:
        raise ValueError("len(maturities) must match hazards.shape[1]")

    # dt per bucket
    dt = torch.empty((num_buckets,), device=device, dtype=FLOAT)
    dt[0] = maturities[0]
    dt[1:] = maturities[1:] - maturities[:-1]

    # survival at endpoints: survival[name, i] = S(T_{i-1}), with survival[:,0]=S(0)=1
    # survival[:, i+1] = survival[:, i] * exp(-hazards[:, i]*dt[i])
    survival = torch.ones((num_names, num_buckets + 1), device=device, dtype=FLOAT)
    survival[:, 1:] = torch.cumprod(torch.exp(-hazards * dt.unsqueeze(0)), dim=1)

    # We need to find i such that survival[name, i] >= u >= survival[name, i+1]
    u = uniforms.T  

    S_i = survival[:, :-1].unsqueeze(2)  
    S_ip1 = survival[:, 1:].unsqueeze(2)  
    u3 = u.unsqueeze(1)                   

    cond = (S_i >= u3) & (u3 >= S_ip1)   
    found = cond.any(dim=1)              

    # bucket index: first True along bucket dimension 
    idx = torch.argmax(cond.to(torch.int64), dim=1)  

    T_im1 = torch.zeros((num_buckets,), device=device, dtype=FLOAT)
    if num_buckets > 1:
        T_im1[1:] = maturities[:-1]

    # Gather per (name, sample)
    idx_g = idx
    lam = hazards.gather(1, idx_g)            
    S_im1 = survival[:, :-1].gather(1, idx_g) 
    Tprev = T_im1.gather(0, idx_g.reshape(-1)).reshape_as(idx_g)  

    # tau = Tprev - (1/lam) * log(u / S_im1)
    # If not found, set tau = maturities[-1] 
    eps = torch.finfo(FLOAT).tiny
    tau_found = Tprev - (1.0 / lam.clamp(min=eps)) * torch.log((u / S_im1).clamp(min=eps))

    tau = torch.where(found, tau_found, maturities[-1].expand_as(tau_found))

    return tau.T  
```

### 3.2 The t-Copula

```python
import scipy
import scipy.special
import scipy.stats

class TCDFPytorch(torch.autograd.Function):
    """
    Forward: SciPy Student-t CDF on CPU (NumPy)
    Backward: exact derivative wrt X using Student-t PDF (SciPy), mapped back to torch
    """
    @staticmethod
    def forward(ctx, X: torch.Tensor, df: float):
        df = float(df)
        ctx.df = df
        ctx.save_for_backward(X)

        X_cpu = X.detach().cpu().numpy()
        U_cpu = scipy.special.stdtr(df, X_cpu)  # CDF
        U = torch.from_numpy(U_cpu).to(device=X.device, dtype=X.dtype)
        return U

    @staticmethod
    def backward(ctx, grad_out: torch.Tensor):
        (X,) = ctx.saved_tensors
        df = ctx.df

        X_cpu = X.detach().cpu().numpy()
        pdf_cpu = scipy.stats.t.pdf(X_cpu, df)  # d/dx CDF = PDF
        pdf = torch.from_numpy(pdf_cpu).to(device=X.device, dtype=X.dtype)

        grad_X = grad_out * pdf
        return grad_X, None  


def t_cdf_autograd(X: torch.Tensor, df: float) -> torch.Tensor:
    return TCDFPytorch.apply(X, df)
```

```python
def t_copula_unif(
    dim: int,
    num_samples: int,
    cov_matrix: torch.Tensor,
    df: float,
    seed: int | None = None,
) -> torch.Tensor:
    """
    Student-t copula with antithetic variates.
    Uses SciPy for the univariate Student-t CDF.

    Returns:
        U: (num_samples, dim) uniforms in (0,1)
    """
    if df <= 0:
        raise ValueError("df must be > 0")
    if num_samples % 2 != 0:
        raise ValueError("num_samples must be even when using antithetic variates")

    g = torch.Generator(device=device)
    if seed is not None:
        g.manual_seed(seed)

    # Cholesky of covariance / correlation matrix
    chol = torch.linalg.cholesky(cov_matrix)

    # iid standard normals (half)
    z = torch.randn(
        num_samples // 2,
        dim,
        generator=g,
        device=device,
        dtype=FLOAT,
    )

    # antithetic normals
    z = torch.cat([z, -z], dim=0)

    # correlate: Y ~ N(0, Σ)
    Y = z @ chol.T

    # Chi-square(df) via Gamma(df/2, 2)
    gamma = torch.distributions.Gamma(
        torch.tensor(df / 2, device=device, dtype=FLOAT),
        torch.tensor(2.0, device=device, dtype=FLOAT),
    )
    S = gamma.sample((num_samples,)).to(dtype=FLOAT)

    # multivariate t
    X = Y / torch.sqrt(S[:, None] / df)

    U = t_cdf_autograd(X, df)

    # numerical safety: keep strictly inside (0,1)
    eps = torch.finfo(FLOAT).eps
    U = U.clamp(min=eps, max=1.0 - eps)

    return U
```

```python
# ---------- parameters ----------
num_samples = 5000
rho_low = 0.0        
rho_corr = 0.9         
rho_anticorr = -0.9  
hazard = 0.02           
# --------------------------------

def compute_2dcov_matrix(rho: float) -> torch.Tensor:
    cov = torch.tensor([[1, rho], [rho, 1]], device=device, dtype=FLOAT)
    return cov

def inverse_exponential(u: torch.Tensor, lam: float) -> torch.Tensor:
    """Inverse CDF of Exp(lam): tau = -log(1-u)/lam"""
    return -torch.log(1.0 - u) / lam

# --- generate uniforms ---
sigma_low = compute_2dcov_matrix(rho_low)
sigma_corr = compute_2dcov_matrix(rho_corr)
sigma_anticorr = compute_2dcov_matrix(rho_anticorr)

U_low       = t_copula_unif(dim=2, num_samples=num_samples, cov_matrix=sigma_low, df=2)
U_corr      = t_copula_unif(dim=2, num_samples=num_samples, cov_matrix=sigma_corr, df=2)
U_anticorr  = t_copula_unif(dim=2, num_samples=num_samples, cov_matrix=sigma_anticorr, df=2)

# --- convert to default times ---
tau_low       = inverse_exponential(U_low,      hazard)
tau_corr      = inverse_exponential(U_corr,     hazard)
tau_anticorr  = inverse_exponential(U_anticorr, hazard)

# --- plot ---
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].scatter(tau_low[:, 0], tau_low[:, 1], s=4, alpha=0.4)
axes[0, 0].set_title("Uncorrelated defaults (ρ = 0)")
axes[0, 0].set_xlabel("τ₁"); axes[0, 0].set_ylabel("τ₂"); axes[0, 0].grid(alpha=0.3)

axes[0, 1].scatter(tau_corr[:, 0], tau_corr[:, 1], s=4, alpha=0.4)
axes[0, 1].set_title("Highly correlated (ρ = 0.9)")
axes[0, 1].set_xlabel("τ₁"); axes[0, 1].set_ylabel("τ₂"); axes[0, 1].grid(alpha=0.3)

axes[1, 0].scatter(tau_anticorr[:, 0], tau_anticorr[:, 1], s=4, alpha=0.4)
axes[1, 0].set_title("Highly anti-correlated (ρ = -0.9)")
axes[1, 0].set_xlabel("τ₁"); axes[1, 0].set_ylabel("τ₂"); axes[1, 0].grid(alpha=0.3)

axes[1, 1].axis('off')  

plt.tight_layout()
plt.show()
```

```python
from enum import Enum

class COPULA_TYPE(Enum):
    GAUSSIAN = 1
    STUDENT_T = 2
    
def compute_copula_unif(
    copula_type: COPULA_TYPE,
    dim: int,
    num_samples: int,
    cov_matrix: torch.Tensor,
    df: float | None = None,
    seed: int | None = None,
) -> torch.Tensor:
    if copula_type == COPULA_TYPE.GAUSSIAN:
        return gaussian_copula_unif(
            dim=dim,
            num_samples=num_samples,
            cov_matrix=cov_matrix,
            seed=seed,
        )
    elif copula_type == COPULA_TYPE.STUDENT_T:
        if df is None:
            raise ValueError("df must be provided for Student-t copula")
        return t_copula_unif(
            dim=dim,
            num_samples=num_samples,
            cov_matrix=cov_matrix,
            df=df,
            seed=seed,
        )
    else:
        raise ValueError("Unsupported copula type")
```

```python
import numpy as np

def probability_of_default(hazards: np.ndarray, tenors: np.ndarray, date: float) -> float:
    """
    hazards[i] applies on (tenors[i-1], tenors[i]] with tenors[-1] >= date.
    Assumes hazards are piecewise-constant across the tenor grid and flat-extended on the last bucket if date > tenors[-1].
    """
    survival_prob, prev = 1.0, 0.0

    for idx, mat in enumerate(tenors):
        if mat <= date:
            dt = mat - prev
            survival_prob *= np.exp(-hazards[idx] * dt)
            prev = mat
        else:
            break
    else:
        # loop didn't break: date is beyond the last tenor; extend last hazard flat
        idx = len(tenors) - 1

    # final stub within the current bucket up to 'date'
    dt = date - prev
    if dt > 0:
        survival_prob *= np.exp(-hazards[idx] * dt)

    return 1.0 - survival_prob  # this is the cumulative PD up to 'date'


print("\nImplied 5Y PD per name:")
for idx, name in enumerate(NAMES):
    hazards = results[idx]["Hazard λ (per year)"]
    PD = probability_of_default(hazards,TENORS,5.0)
    print(f"{name:25s}  PD5Y = {PD*100:.2f}%")
```

```python
def accrual_on_default(last_coupon_t: torch.Tensor, default_t: torch.Tensor) -> torch.Tensor:
    return torch.clamp(default_t - last_coupon_t, min=0.0)


def discount_at(t: torch.Tensor, payment_days: torch.Tensor, discounts: torch.Tensor) -> torch.Tensor:
    """
    Vectorized piecewise-linear interpolation matching your Python helper.
    """
    below = t <= payment_days[0]
    above = t >= payment_days[-1]

    idx = torch.searchsorted(payment_days, t, right=False)  # first i with payment_days[i] >= t
    idx1 = idx.clamp(min=0, max=payment_days.numel() - 1)
    idx0 = (idx1 - 1).clamp(min=0, max=payment_days.numel() - 1)

    t0, t1 = payment_days[idx0], payment_days[idx1]
    d0, d1 = discounts[idx0], discounts[idx1]

    denom = (t1 - t0).clamp(min=1e-18)
    w = (t - t0) / denom
    out = d0 + w * (d1 - d0)

    out = torch.where(below, discounts[0].expand_as(out), out)
    out = torch.where(above, discounts[-1].expand_as(out), out)
    return out
```

```python
def price_kth_to_default_cds(
    k: int,
    recovery: float,
    maturities: np.ndarray,
    payment_days: np.ndarray,
    discounts: np.ndarray,
    hazards: np.ndarray,
    cov_matrix: np.ndarray,
    num_paths: int = 20000,
    copula_type: COPULA_TYPE = COPULA_TYPE.GAUSSIAN,
    df: float = 3.0,
    seed: int | None = 42,
) -> dict[str, float]:
    """
    Computes fair spread of k-th to default basket CDS (fully vectorized using Pytorch).
    Using following conventions:
      - AoD: max(tau_k - last_coupon, 0)
      - discount_at: piecewise-linear interp
      - uniforms_to_default_times: returns maturities[-1] when not found (no default by last bucket)
      - antithetic Gaussian copula sampling
    """
    maturities_t = torch.as_tensor(maturities, device=device, dtype=FLOAT)
    payment_days_t = torch.as_tensor(payment_days, device=device, dtype=FLOAT)
    discounts_t = torch.as_tensor(discounts, device=device, dtype=FLOAT)
    hazards_t = torch.as_tensor(hazards, device=device, dtype=FLOAT)
    cov_t = torch.as_tensor(cov_matrix, device=device, dtype=FLOAT)

    # time deltas
    deltas_t = torch.empty_like(payment_days_t)
    deltas_t[0] = payment_days_t[0]
    deltas_t[1:] = payment_days_t[1:] - payment_days_t[:-1]

    # sample & invert
    U = compute_copula_unif(
        copula_type=copula_type,
        dim=hazards_t.shape[0],
        num_samples=num_paths,
        cov_matrix=cov_t,
        df=df,
        seed=seed,
    )
    tau = uniforms_to_default_times(U, hazards_t, maturities_t)  

    # kth order statistic
    tau_sorted, _ = torch.sort(tau, dim=1)
    tau_k = tau_sorted[:, k - 1]
    T = payment_days_t[-1]

    # protection leg
    df_tau = discount_at(tau_k, payment_days_t, discounts_t)
    prot_cf = torch.where(tau_k <= T, (1.0 - recovery) * df_tau, torch.zeros_like(df_tau))
    pv_prot = prot_cf.mean()

    # premium PV01
    full_pv01 = torch.sum(deltas_t * discounts_t)

    idx = torch.searchsorted(payment_days_t, tau_k, right=False)  # first >= tau_k
    idxm1 = (idx - 1).clamp(min=-1)

    csum = torch.cumsum(deltas_t * discounts_t, dim=0)
    coupons_pv = torch.zeros_like(tau_k)
    has_coupon = idx > 0
    coupons_pv[has_coupon] = csum[idxm1[has_coupon]]

    last_coupon_t = torch.zeros_like(tau_k)
    last_coupon_t[has_coupon] = payment_days_t[idxm1[has_coupon]]

    accr = accrual_on_default(last_coupon_t, tau_k)
    pv01 = coupons_pv + accr * df_tau

    # if tau_k > T: full schedule
    pv01 = torch.where(tau_k > T, full_pv01.expand_as(pv01), pv01)

    prem_cf = pv01 / 1e4
    pv_prem = prem_cf.mean()

    stderr = prot_cf.std(unbiased=True) / torch.sqrt(torch.tensor(float(num_paths), device=device, dtype=FLOAT))
    fair_s_bps = 1e4 * (pv_prot / (pv01.mean() + 1e-30))

    return {
        "pv_protection": float(pv_prot.item()),
        "pv_premium_at_coupon_bps": float(pv_prem.item()),
        "fair_spread_bps": float(fair_s_bps.item()),
        "std_error": float(stderr.item()),
    }
```
